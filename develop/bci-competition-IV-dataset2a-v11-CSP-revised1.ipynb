## Decompose composite covariance

```python
sigma_total = sigma0 + sigma1
```

```python
eigval_c, eigvec_c = np.linalg.eig(sigma_total)
```

```python
idx_dsc = np.argsort(eigval_c)[::-1]
```

```python
eigval_c_desc = eigval_c[idx_dsc]
```

```python
eigval_c_desc
```

```python
eigval_c_desc = np.diag(eigval_c_desc)
```

```python
eigvec_c_desc = eigvec_c[:, idx_dsc]
```

## White matrix

```python
P = np.sqrt(np.linalg.inv(eigval_c_desc))@eigvec_c_desc.T
```

```python
def ifWhiteIdentity(P, Cov_c):
    '''
    '''
    # Try if eigenval of PCcP' equal to identity matrix
    temp = P@Cov_c@P.T
    temp = np.diag(np.linalg.eigvals(temp))
    
    return abs(temp)
```
